```python
from pathlib import Path
import frontmatter

from langchain.schema import Document
from langchain.text_splitter import (
    RecursiveCharacterTextSplitter,
)
from langchain_community.document_loaders import (
    DirectoryLoader,
    TextLoader,
)

# Path to Docker docs
DOCS_PATH = Path(
    "../data/docs/external/docker-docs/content/get-started"
)


def load_markdown_docs():
    """Load raw markdown/mdx files."""

    documents = []

    # Load both .md and .mdx files
    for pattern in ("**/*.md", "**/*.mdx"):

        loader = DirectoryLoader(
            str(DOCS_PATH),
            glob=pattern,
            loader_cls=TextLoader,
            loader_kwargs={"encoding": "utf-8"},
            show_progress=True,
        )

        documents.extend(loader.load())

    return documents


def clean_documents(documents):
    """Remove frontmatter and keep useful metadata."""

    cleaned_documents = []

    for doc in documents:

        # Parse frontmatter
        post = frontmatter.loads(doc.page_content)

        # Main markdown content
        cleaned_content = post.content

        # Create cleaned document
        cleaned_doc = Document(
            page_content=cleaned_content,
            metadata={
                "source": doc.metadata.get("source", ""),
                "title": post.metadata.get("title", ""),
                "description": post.metadata.get(
                    "description", ""
                ),
            },
        )

        cleaned_documents.append(cleaned_doc)

    return cleaned_documents


def chunk_documents(documents):
    """Split documents into smaller chunks."""

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=700,      # Max chunk size
        chunk_overlap=100,   # Shared context
        length_function=len,
    )

    chunks = splitter.split_documents(documents)

    return chunks


if __name__ == "__main__":

    # Load raw markdown docs
    raw_documents = load_markdown_docs()

    print(f"\nLoaded documents: {len(raw_documents)}")

    # Clean markdown content
    cleaned_documents = clean_documents(raw_documents)

    print(
        f"Cleaned documents: {len(cleaned_documents)}"
    )

    # Split into chunks
    chunks = chunk_documents(cleaned_documents)

    print(f"Total chunks: {len(chunks)}")

    # Preview first chunk
    first_chunk = chunks[0]

    print("\nChunk Metadata:")
    print(first_chunk.metadata)

    print("\nChunk Content:")
    print(first_chunk.page_content[:700])
```
